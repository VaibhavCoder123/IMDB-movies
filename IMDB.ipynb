 # So Many errors

# Lets try to remove errors(NaN values)

I have made 'Year' column clear here

```python
dataframe['Year'] = pd.to_numeric(dataframe['Year'], errors='coerce')
```

# Here we will make a new variable which is tdf ( True Dataframe )

```python
true_dataframe = dataframe.dropna()
```

# I removed the error by using [dropna]

```python
true_dataframe
```

```python
sns.heatmap(true_dataframe.isnull())
```

```python
true_dataframe.isnull().sum()
```

```python
true_dataframe.shape
```

```python
true_dataframe.info()
```

# Now we will check that are there any duplicate values or not

```python
dup_true_dataframe=true_dataframe.duplicated().any()
```

```python
print("is there any duplicate value? : ",dup_true_dataframe)
```

# This means our dataset is clear now

# I am making another variable for this   [ddf(duplicated dataframe)]

```python
duplicate_dataframe = true_dataframe.drop_duplicates()
duplicate_dataframe
```

```python
duplicate_dataframe.describe()
```

```python
duplicate_dataframe.describe(include='all')
```

# Our data is clear now

```python
clear_data = duplicate_dataframe
```

# Now movies which are less than 90 minute

```python
duplicate_dataframe[duplicate_dataframe['Duration']<=90]
```

This shows that total 23 movies are less than 90 minute

# Lets check for greater than 200 minute

```python
duplicate_dataframe[duplicate_dataframe['Duration']>=200]
```

This shows that total 1 movie is equal or greater than 200 minute

# This means that 204 movies are between 90 and 200 minuts

# Lets find out that how many movies were made between 1940 to 2023

```python
duplicate_dataframe[duplicate_dataframe['Year']<=1940]
```

This shows that the oldest movie was made in 1938

# Now we will make a graph

```python
w = clear_data.Votes
x = clear_data.No_of_movies
y = clear_data.Rating
z = clear_data.Duration
```

```python
plt.scatter(x,y,marker='*',c='cyan',ec='red',alpha=0.7)
plt.xticks(rotation=60)
plt.xlabel("No. of Movies",fontsize='30')
plt.ylabel("Ratings",fontsize="30")
plt.show()
```

# Now Line Plots

```python
plt.plot(x,y,label='Ratings',c='red',lw=2)
plt.plot(x,z,label='Duration',c='blue',lw=2)
plt.xlabel('No. of Movies',fontsize='20')
plt.ylabel("Ratings,votings and duration of movies",fontsize='10')
plt.xticks (rotation=30)
plt.yticks (rotation=30)
plt.legend()
plt.show()
```

```python
plt.plot(x,w,label='Votes',c='red',alpha=0.7,lw=2)
plt.xlabel('No. of Movies',fontsize='20')
plt.ylabel("Votes of movies",fontsize='10')
plt.xticks (rotation=30)
plt.yticks (rotation=30)
plt.legend()
plt.show()
```

# Now Histogram of Ratings

```python
plt.hist(y,bins=30, color="cyan",ec="red", lw=2,)
plt.title("Histogram of Ratings", fontsize=30)
plt.xlabel("Ratings", fontsize=20)
plt.ylabel("No. of Movies",fontsize=20)
plt.show()
```

# Now Histogram of Duration

```python
plt.hist(z,bins=30, color="green",ec="pink", lw=2,)
plt.title("Histogram of Duration", fontsize=30)
plt.ylabel("No. of Movies",fontsize=20)
plt.xlabel("Duration of movies", fontsize=20)
plt.show()
```

# Now Histogram of Votes

```python
plt.hist(w,bins=30, color="green",ec="pink", lw=2,)
plt.title("Histogram of Votes", fontsize=30)
plt.ylabel("No. of Movies",fontsize=20)
plt.xlabel("Votes of movies", fontsize=20)
plt.show()
```
